# Приклад Інкапсуляції в Python

- прграма яка визначає чи студент занудбгував на лекції

```python
class CheckSudentBored:
    def __init__(self, name, words_threshhold=50):
        self.__name = name          # Приватний атрибут
        self.__words_threshhold = words_threshhold  # Приватний атрибут
    
    def __generate_lecture_words(self):
        # Приватний метод для генерації кількості слів
        import random
        return random.randint(0, 100)
    
    def __is_bored(self):
        lecture_words = self.__generate_lecture_words()
        return lecture_words > self.__words_threshhold
    
    def check_if_student_bored(self):
        if self.__is_bored():
            return f"Студент {self.__name} занудьгував на лекції."
        else:
            return f"Студент {self.__name} НЕ занудьгував на лекції."

b = CheckSudentBored("Віктор", words_threshhold=40)
b.check_if_student_bored()
```

# Приклад наслідування в Python
- напишемо програму яка імітує роботу станції беперебійного живлення
- ми розширимо дану стінцію щоб можна було генерувати елктрику від сонячних панелей, вітряків та дизельного генератора
- альтернативно ми придумаємо футуристичні джерела енергії

```python
class EnergyDevice:
    def __init__(self, name, capacity):
        self.name = name
        self.capacity = capacity
        self.__max_capacity = capacity

    @property
    def info(self):
        return f"Пристрій: {self.name}, Ємність: {self.__max_capacity}"
    
    def recalculate_total_capacity(self, additional_capacity=0):
        self.__max_capacity += additional_capacity

    def charge_device(self):
        print(f"{self.name} заряджається до ємності {self.capacity}.")
        self.capacity += 10 if self.capacity < self.__max_capacity else self.__max_capacity
        return self.capacity
    
    def generate_energy(self):
        print(f"{self.name} генерує енергію до ємності {self.capacity}.")
        self.capacity -= 10 if self.capacity > 0 else 0
        return self.capacity

E = EnergyDevice("Батарея", 100)
E.generate_energy()
E.generate_energy()
E.generate_energy()
E.charge_device()
E.charge_device()
print(E.info)
```

```python
class SolarPanelDevice(EnergyDevice):
    def __init__(self, name, capacity, panel_area):
        super().__init__(name, capacity)
        self.panel_area = panel_area
        self.recalculate_total_capacity(self.__calculate_additional_capacity)
    
    @property
    def __calculate_additional_capacity(self):
        return self.panel_area * 2  # Припустимо, що кожен квадратний метр додає 2 одиниці ємності
    
    @property
    def info(self):
        base_info = super().info
        return f"{base_info}, Площа панелі: {self.panel_area} м²"

S = SolarPanelDevice("Сонячна Панель", 100, 20)
print(S.info)
S.charge_device()
S.generate_energy()
S.generate_energy()
```

```python
class HybridSolarWindDevice(SolarPanelDevice):
    def __init__(self, name, capacity, panel_area, wind_turbine_power):
        super().__init__(name, capacity, panel_area)
        self.wind_turbine_power = wind_turbine_power
        self.recalculate_total_capacity(self.__calculate_wind_additional_capacity)
        self.Test = "Hello"
    
    @property
    def __calculate_wind_additional_capacity(self):
        return self.wind_turbine_power * 3  # Припустимо, що кожен кіловат додає 3 одиниці ємності
    
    @property
    def info(self):
        base_info = super().info
        return f"{base_info}, Потужність вітряка: {self.wind_turbine_power} кВт"

H = HybridSolarWindDevice("Гібридна Сонячно-Вітряна Система", 150, 30, 5)
print(H.info)
H.generate_energy()
H.charge_device()
```

```python
class TeslaEtheriumDevice:
    def __init__(self, name, etherium_level):
        self.name = name
        self.capacity = etherium_level * 0.2 # Припустимо, що кожен рівень ефіріуму додає 0.2 одиниці ємності

class AdvancedTeslaDevice(HybridSolarWindDevice, TeslaEtheriumDevice):
    def __init__(self, name, capacity, panel_area, wind_turbine_power, etherium_level):
        HybridSolarWindDevice.__init__(self, name, capacity, panel_area, wind_turbine_power)
        TeslaEtheriumDevice.__init__(self, name, etherium_level)
        self.recalculate_total_capacity(self.__calculate_etherium_additional_capacity)
    
    @property
    def __calculate_etherium_additional_capacity(self):
        return self.capacity  # Використовуємо ємність з TeslaEtheriumDevice як додаткову ємність
    
    @property
    def info(self):
        base_info = super().info
        return f"{base_info}, Рівень Ефіріуму: {self.capacity / 0.2}"

A = AdvancedTeslaDevice("Просунута Тесла Система", 200, 40, 10, 50)
A.Hello = 10
print(A.Test, A.Hello)
```
