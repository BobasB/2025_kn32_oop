# Приклад Інкапсуляції в Python

- прграма яка визначає чи студент занудбгував на лекції

```python
class CheckSudentBored:
    def __init__(self, name, words_threshhold=50):
        self.__name = name          # Приватний атрибут
        self.__words_threshhold = words_threshhold  # Приватний атрибут
    
    def __generate_lecture_words(self):
        # Приватний метод для генерації кількості слів
        import random
        return random.randint(0, 100)
    
    def __is_bored(self):
        lecture_words = self.__generate_lecture_words()
        return lecture_words > self.__words_threshhold
    
    def check_if_student_bored(self):
        if self.__is_bored():
            return f"Студент {self.__name} занудьгував на лекції."
        else:
            return f"Студент {self.__name} НЕ занудьгував на лекції."

b = CheckSudentBored("Віктор", words_threshhold=40)
b.check_if_student_bored()
```

# Приклад наслідування в Python
- напишемо програму яка імітує роботу станції беперебійного живлення
- ми розширимо дану стінцію щоб можна було генерувати елктрику від сонячних панелей, вітряків та дизельного генератора
- альтернативно ми придумаємо футуристичні джерела енергії

```python
class EnergyDevice:
    def __init__(self, name, capacity):
        self.name = name
        self.capacity = capacity
        self.__max_capacity = capacity

    @property
    def info(self):
        return f"Пристрій: {self.name}, Ємність: {self.__max_capacity}"
    
    def recalculate_total_capacity(self, additional_capacity=0):
        self.__max_capacity += additional_capacity

    def charge_device(self):
        print(f"{self.name} заряджається до ємності {self.capacity}.")
        self.capacity += 10 if self.capacity < self.__max_capacity else self.__max_capacity
        return self.capacity
    
    def generate_energy(self):
        print(f"{self.name} генерує енергію до ємності {self.capacity}.")
        self.capacity -= 10 if self.capacity > 0 else 0
        return self.capacity

E = EnergyDevice("Батарея", 100)
E.generate_energy()
E.generate_energy()
E.generate_energy()
E.charge_device()
E.charge_device()
print(E.info)
```

```python
class SolarPanelDevice(EnergyDevice):
    def __init__(self, name, capacity, panel_area):
        super().__init__(name, capacity)
        self.panel_area = panel_area
        self.recalculate_total_capacity(self.__calculate_additional_capacity)
    
    @property
    def __calculate_additional_capacity(self):
        return self.panel_area * 2  # Припустимо, що кожен квадратний метр додає 2 одиниці ємності
    
    @property
    def info(self):
        base_info = super().info
        return f"{base_info}, Площа панелі: {self.panel_area} м²"

S = SolarPanelDevice("Сонячна Панель", 100, 20)
print(S.info)
S.charge_device()
S.generate_energy()
S.generate_energy()
```

```python
class HybridSolarWindDevice(SolarPanelDevice):
    def __init__(self, name, capacity, panel_area, wind_turbine_power):
        super().__init__(name, capacity, panel_area)
        self.wind_turbine_power = wind_turbine_power
        self.recalculate_total_capacity(self.__calculate_wind_additional_capacity)
        self.Test = "Hello"
    
    @property
    def __calculate_wind_additional_capacity(self):
        return self.wind_turbine_power * 3  # Припустимо, що кожен кіловат додає 3 одиниці ємності
    
    @property
    def info(self):
        base_info = super().info
        return f"{base_info}, Потужність вітряка: {self.wind_turbine_power} кВт"

H = HybridSolarWindDevice("Гібридна Сонячно-Вітряна Система", 150, 30, 5)
print(H.info)
H.generate_energy()
H.charge_device()
```

```python
class TeslaEtheriumDevice:
    def __init__(self, name, etherium_level):
        self.name = name
        self.capacity = etherium_level * 0.2 # Припустимо, що кожен рівень ефіріуму додає 0.2 одиниці ємності

class AdvancedTeslaDevice(HybridSolarWindDevice, TeslaEtheriumDevice):
    def __init__(self, name, capacity, panel_area, wind_turbine_power, etherium_level):
        HybridSolarWindDevice.__init__(self, name, capacity, panel_area, wind_turbine_power)
        TeslaEtheriumDevice.__init__(self, name, etherium_level)
        self.recalculate_total_capacity(self.__calculate_etherium_additional_capacity)
    
    @property
    def __calculate_etherium_additional_capacity(self):
        return self.capacity  # Використовуємо ємність з TeslaEtheriumDevice як додаткову ємність
    
    @property
    def info(self):
        base_info = super().info
        return f"{base_info}, Рівень Ефіріуму: {self.capacity / 0.2}"

A = AdvancedTeslaDevice("Просунута Тесла Система", 200, 40, 10, 50)
A.Hello = 10
print(A.Test, A.Hello)
```

# Приклад поліморфізму в Python
- першим поліморфізмом є перевантаження методів через різну кількість аргументів і їх типів
- другим - поліморфізм з наслідуванням, коли підкласи мають методи з однаковими іменами, але різною реалізацією
- третій Duck Typing - коли об'єкти різних класів можуть використовуватися взаємозамінно, якщо вони реалізують однаковий інтерфейс або методи
- напишемо програму яка імітує роботу Павербанку, його зярядка від різних пристроїв та його взаємодію з різними пристроями 

```python
class PowerBank:
    def __init__(self, name, capacity):
        self.name = name
        self.capacity = capacity
        self.__max_capacity = capacity
    
    def charge_device_from(self, device):
        print(f"{self.name} заряджається від {device.name}.")
        device.charge_device(self.capacity, self.__max_capacity)

class OutletDevice:
    def __init__(self, name):
        self.name = name
    
    def charge_device(self, powerbank_capacity, powerbank_max_capacity):
        print(f"{self.name} заряджає пристрій з ємністю {powerbank_capacity}.")
        powerbank_capacity -= 20 if powerbank_capacity > 0 else 0
        return powerbank_capacity
    
class USBDevice:
    def __init__(self, name):
        self.name = name
    
    def charge_device(self, powerbank_capacity, powerbank_max_capacity):
        print(f"{self.name} заряджає пристрій з ємністю {powerbank_capacity}.")
        powerbank_capacity -= 10 if powerbank_capacity > 0 else 0
        return powerbank_capacity

class BrokenChargerDevice:
    def __init__(self, name):
        self.model = name
    
    def charge(self, powerbank_capacity, powerbank_max_capacity):
        print(f"{self.model} не може заряджати пристрій через поломку.")
    

# Використання поліморфізму типу "Duck Typing"
P = PowerBank("Мій Павербанк", 100)
O = OutletDevice("Розетка")
U = USBDevice("USB Пристрій")
B = BrokenChargerDevice("Зламаний Зарядний Пристрій")

P.charge_device_from(O)
P.charge_device_from(U)
#P.charge_device_from(B)  # Це викличе помилку, оскільки метод називається інакше
```

```python
# В Пайтон неможна створювати функції з однаковими іменами, оскільки остання визначена функція перезаписує попередні.
def example_function():
    return "Перша версія функції"

def example_function():
    return "Друга версія функції"
# в Пафтон, це не є поліморфізмом, оскільки перша функція буде втрачена
print(example_function())

# поліморфізм через дефолтні аргументи
def example_function_with_args(arg1, arg2=0):
    if arg2 == 0:
        return f"Функція з одним аргументом: {arg1}"
    return f"Функція з аргументами: {arg1}, {arg2}"

print(example_function_with_args("Тест"), example_function_with_args("Тест", 5))

# поліморфізм через змінну кількість аргументів та ключові аргументи
def example_function_var_args(*args, **kwargs):
    result = "Функція з змінною кількістю аргументів:\n"
    for i, arg in enumerate(args):
        result += f"  Аргумент {i+1}: {arg}\n"
    for key, value in kwargs.items():
        result += f"  Ключовий аргумент {key}: {value}\n"
    return result

print(example_function_var_args(1, 2, 3, name="Тест", value=10))
print(example_function_var_args("Один аргумент"))
print(example_function_var_args())
```

# Приклад Абстракції в Python

- подивимось абстрацію на прикладі LLM  моделі
- спочатку вторимо абстрактний клас LLMModel з абстрактними методами askLLm та genereateResponse
- далі спробуємо реалізувати протокол через typing.Protocol для LLMModel

```python
from abc import ABC, abstractmethod

class LLMModel(ABC):
    @abstractmethod
    def askLLm(self, question):
        pass
    
    @abstractmethod
    def generateResponse(self, prompt):
        pass

class ItCollegeLLMModel(LLMModel):
    def __init__(self, model_name):
        self.model_name = model_name
    
    def askLLm(self, question):
        return f"{self.model_name} відповідає на питання: {question}"
    
    def generateResponse(self, prompt):
        return f"{self.model_name} генерує відповідь на запит: {prompt}"

class VoloymyrVLLMModel(LLMModel):
    def __init__(self, model_name):
        self.model_name = model_name
    
    def askLLm(self, question):
        return f"{self.model_name} відповідає на питання: {question}"
    
    def generateResponse(self, prompt):
        return f"{self.model_name} генерує відповідь на запит: {prompt}"
    

itllm = ItCollegeLLMModel("IT College LLM")
print(itllm.askLLm("Як працює LLM?"))
print(itllm.generateResponse("Які є переваги LLM?"))

vvmodel = VoloymyrVLLMModel("Voloymyr V LLM")
print(vvmodel.askLLm("Що таке поліморфізм?"))
print(vvmodel.generateResponse("Як реалізувати поліморфізм в Python?"))
```

```python
from typing import Protocol
from abc import ABC, abstractmethod

class LLMModelProtocol(Protocol):
    @property
    def model_name(self) -> str:
        ...
    
    def tokenize(self, text: str) -> list:
        ...

class ModelInterface(ABC):
    @abstractmethod
    def askLLm(self, question):
        pass
    
    @abstractmethod
    def generateResponse(self, prompt):
        pass

class GPT3Model(LLMModelProtocol):
    def __init__(self, model_name):
        self._model_name = model_name
    
    @property
    def model_name(self) -> str:
        return self._model_name
    
    def tokenize(self, text: str) -> list:
        return text.split()  # Простий приклад токенізації

class BERTModel(LLMModelProtocol):
    def __init__(self, model_name):
        self._model_name = model_name
    
    @property
    def model_name(self) -> str:
        return self._model_name
    
    def tokenize(self, text: str) -> list:
        return text.split()  # Простий приклад токенізації

class ChatInterface(ModelInterface):
    def __init__(self, llm_model: LLMModelProtocol):
        self.llm_model = llm_model
    
    def askLLm(self, question):
        tokens = self.llm_model.tokenize(question)
        return f"{self.llm_model.model_name} відповідає на питання: {tokens}"
    
    def generateResponse(self, prompt):
        tokens = self.llm_model.tokenize(prompt)
        return f"{self.llm_model.model_name} генерує відповідь на запит: {tokens}"


for model in [GPT3Model("GPT-3"), BERTModel("BERT")]:
    chat_interface = ChatInterface(model)
    print(chat_interface.askLLm("Як працює LLM?"))
    print(chat_interface.generateResponse("Які є переваги LLM?"))
```
