Створити клас ProductCalculator, який:
- Приймає в конструкторі список чисел
- Має метод calculate_product(), який обчислює добуток всіх елементів списку
- Має property product, яка повертає добуток всіх чисел

Не потрібно пояснювати код. Використай форматування Markdown для вставки коду. У функції print використовуй українську мову.

```python
from functools import reduce
from typing import List


class ProductCalculator:
    def __init__(self, numbers: List[float]):
        self._numbers = numbers
        self._product = None

    def calculate_product(self) -> float:
        self._product = reduce(lambda x, y: x * y, self._numbers, 1)
        return self._product

    @property
    def product(self) -> float:
        if self._product is None:
            self.calculate_product()
        return self._product

    def __str__(self) -> str:
        return "Ім'я: Богдан, Група: КН-32"


data_sets = [
    [1, 2, 3, 4],
    [2, 5, 10],
    [0.5, 4, 8]
]

calculators = []

for data in data_sets:
    calculator = ProductCalculator(data)
    calculators.append(calculator)
    print(str(calculator))
    print(f"Вхідні дані: {data}")
    print(f"Добуток чисел: {calculator.product}")
    print("-" * 40)
```
